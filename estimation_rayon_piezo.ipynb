```python
import numpy
import time, os, sys
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from tifffile import imwrite
from matplotlib.pyplot import figure

```

```python
vjet=14.6    #vitesse du jet
Rjet=100e-6    #rayon du jet 
Rdrop_static=1.89*Rjet    #rayon théorique en sortie de jet

print(Rdrop_static*1e6)
```

```python
fexc=np.linspace(1,25000,1000)     #vecteur des fréquences
Rdrop=((3*(Rjet**2)*vjet)/(4*fexc))**(1/3)     #formule du rayon des gouttes en fonction de la fréquence piezo
Rdrop_static2=np.linspace(1,1,1000)*Rdrop_static    
```

Plot du rayon des gouttes en fonction de la fréquence piezo

```python
plt.plot(fexc,Rdrop*1e6)
plt.ylabel('Drop radius (µm)')
plt.xlabel('Frequency (Hz)')
plt.plot(fexc,Rdrop_static2*1e6)
plt.ylim([100,400])
plt.xlim([0,20000])
plt.legend(['Drop size piezo','Without'])
```
