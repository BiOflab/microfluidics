```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

import numpy
import time, os, sys
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from tifffile import imwrite
from matplotlib.pyplot import figure
from scipy.special import kn
from scipy.special import yn


```

```python

def relation_dispersion(N,Gamma,Oh):
    oh=Oh
    XX=np.linspace(0,2,N)
    k2=np.zeros(N)
    gamma=Gamma

    for ii in range(N):
        x=XX[ii]
        a=1
        b=-x**2*(24+x**2)/(8+x**2)*oh
        c=-4*x**2/(8+x**2)*(1-x**2-gamma*(1-x*kn(1,x)/kn(0,x)))

        delta=b*b-4*a*c
        if delta>=0:
            solution=(b+np.sqrt(delta))/2/a
            k2[ii]=solution

            
    return XX, k2
```

EFFET CHAMP ELECTRIQUE

```python
R=100e-6
tension_surface=55e-3
rho=1000
w0=np.sqrt(tension_surface/rho/R**3)
f0=1/(2*3.14/1.46*100e-6)


xx, kk=relation_dispersion(100,0.,0.3)
xx2, kk2=relation_dispersion(100,0.5,0.3)

plt.plot(xx,kk)
plt.plot(xx2,kk2)

plt.ylim([0, .25])
plt.xlim([0, 1.25])
```
