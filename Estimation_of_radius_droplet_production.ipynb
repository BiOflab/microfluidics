# ESTIMATE THE RADIUS OF CAPSULES FALLING / SORBITOL CONDITION

IMPORT LIBRAIRIES AND VIDEO

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from scipy.signal import find_peaks
import sys
import cv2 as cv
im = io.imread('sorbitol.tif')
im=255-im
```

 DISPLAY A GIVEN FRAME

```python
plt.imshow(im[1950,:,:], cmap='gray')
```

SELECT A POSITION AND DETECT THE CIRCLE / DROPLET

```python
test=im[4030,60:120,400:550]
plt.imshow(test)

circles = cv2.HoughCircles(test, cv2.HOUGH_GRADIENT, 1.2, 3,
                           param1 = 40, param2 = 22, minRadius = 5, maxRadius = 25)
        
    
    
if type(circles) is not np.ndarray :                                   
    print('tutut')                                   
elif type(circles) is np.ndarray :
    circles = np.uint16(np.around(circles))     
    for pt in circles[0,:]:                    
        q,s,d = pt[0],pt[1],pt[2]
        print(d)
        cv2.circle(test,(q,s),d,(0,0,255),1)
        plt.imshow(test)
```

RUN THE DETECTION FOR THE ENTIRE VIDEO

```python
count=0

RR=np.zeros(100000)

for ii in range(im.shape[0]):
    test=im[ii,60:120,400:550]
    circles = cv2.HoughCircles(test, cv2.HOUGH_GRADIENT, 1.2, 3,
                           param1 = 40, param2 = 22, minRadius = 3, maxRadius = 25)    
        
    if type(circles) is not np.ndarray :                                   
        print('rat')                                   
    elif type(circles) is np.ndarray :
        circles = np.uint16(np.around(circles))     
        for pt in circles[0,:]:
            count=count+1
            q,s,d = pt[0],pt[1],pt[2]
            RR[count]=d
```

COMPUTE THE MEAN DIAMETER AND DISPLAY THE HISTOGRAM

```python
RR=RR[0:count]
np.mean(RR)*2*13.7
n_bins=25
plt.hist(RR*2*13.7,bins=n_bins)
plt.title("Intensity histogram")
plt.xlabel('Diameter (m)')
plt.ylabel('Count (a.u)')
print('mean diameter =',np.mean(RR)*2*13.7)
```
